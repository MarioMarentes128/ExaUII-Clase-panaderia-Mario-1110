```python
class Panaderia:
    def __init__(self, nombre, direccion, especialidad):
        # Atributos de la clase
        self.nombre = nombre
        self.direccion = direccion
        self.especialidad = especialidad

    def leerdatos(self):
        # Función que confirma la lectura de los datos
        return f"El sistema ha registrado exitosamente a: '{self.nombre}'."

    def mostrardatos(self):
        # Función que muestra la información al usuario
        print(f"\n--- Ficha Informativa ---")
        print(f"Nombre: {self.nombre}")
        print(f"Dirección: {self.direccion}")
        print(f"Especialidad: {self.especialidad}")
        print("-------------------------\n")

# --- Bloque interactivo para capturar datos ---

print("=== Registro de Panadería: Local 1 ===")
n1 = input("Nombre: ")
d1 = input("Dirección: ")
e1 = input("Especialidad: ")
Local1 = Panaderia(n1, d1, e1)

print("\n=== Registro de Panadería: Local 2 ===")
n2 = input("Nombre: ")
d2 = input("Dirección: ")
e2 = input("Especialidad: ")
Local2 = Panaderia(n2, d2, e2)

# --- Ejecución de las funciones para ambos objetos ---

# Interacción con Local 1
print(f"\n[Sistema] {Local1.leerdatos()}")
Local1.mostrardatos()

# Interacción con Local 2
print(f"[Sistema] {Local2.leerdatos()}")
Local2.mostrardatos()

print("Programa realizado por Mario Yañez NL = 53")
```
